```python
from bs4 import BeautifulSoup
```

```python
url = "http://earthsendangered.com/search-regions3.asp?search=1&sgroup=allgroups&ID=44"
```

```python
page = requests.get(url)

#parse the html

animals = BeautifulSoup(page.content, 'html.parser') 
```

```python
animals
```

```python
animals.find_all("table",class_="table table-bordered table-striped")
```

```python
length = len(animals.find_all("table",class_="table table-bordered table-striped")[0].find_all("a"))
animal_list = []
for i in range(length):
    animal_name = animals.find_all("table",class_="table table-bordered table-striped")[0].find_all("a")[i].get_text()
    animal_list.append(animal_name)
```

```python
animal_list #common name and needs to be cleaned
```

```python
animals.find_all("td", style="vertical-align: top;")
```

```python
#len(animals.find_all("td", style="vertical-align: top;"))
name_list = []
scientific_list = []
species_list = []
countries_list = []
for i in range(0,len(animals.find_all("td", style="vertical-align: top;")),4):
    common_name = animals.find_all("td", style="vertical-align: top;")[i].get_text()
    scientific_name = animals.find_all("td", style="vertical-align: top;")[i+1].get_text()
    species = animals.find_all("td", style="vertical-align: top;")[i+2].get_text()
    country = animals.find_all("td", style="vertical-align: top;")[i+3].get_text()
    name_list.append(common_name)
    scientific_list.append(scientific_name)    
    species_list.append(species) 
    countries_list.append(country)
```

```python
name_list
```

```python
import re
name_list_clean = []
for name in name_list:
    name_clean = name.strip()
    name_list_clean.append(name_clean)
```

```python
name_list_clean
```

```python
scientific_list
```

```python
species_list
```

```python
countries_list
```

```python
#let's create the dataframe with the data retrieved

EndangeredAnimals = pd.DataFrame({"Common_name":name_list_clean,"Scientific_name":scientific_list,"Species_group":species_list,"Country":countries_list})
```

```python
EndangeredAnimals
```

```python
EndangeredAnimals.to_csv("Brasil_Endangered_Species.csv",index=False)
```
