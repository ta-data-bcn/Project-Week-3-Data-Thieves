### Web Scrapping Wikipedia page with relevant MWC events

```python
import requests
import re

url = 'https://es.wikipedia.org/wiki/Mobile_World_Congress'
html = requests.get(url).content
html[0:600]
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(html, "lxml")
soup
```

#### RegEx method to isolate important dates

```python
tags = ['li']
text_wiki = [element.text for element in soup.find_all(tags)]
text_wiki
```

```python
type(text_wiki)
```

```python
year = '\d{4}'
```

```python
testo = '\d{1,2}\s\w+\s\d{1,2}\s\w+\s.+?\s|\d{1,2}\s\w+\s\w+\s\w+\s\d{1,2}.+?\.'
```

```python
lista = []
for i in text_wiki:
    x = re.findall(testo, i)
    lista.append(x)
```

```python
lista2 = [x for x in lista if x != []]
```

```python
lista2
```

```python
for i in lista2:
    lista2.str_replace('de febrero','/2')
    lista2
```

```python
for x in lista2:
    map(lambda x: str.replace(x, 'de febrero','/2'), lista2)
```
