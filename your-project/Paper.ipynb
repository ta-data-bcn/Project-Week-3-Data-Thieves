**The End of YouTube As We Know It? An Exploratory Analysis of How the Algorithm Change Impacted on Content Creators**


***Abstract***

<div class=text-justify>
Artifical intelligence driven recommendations are responsbile for over 70 percent of all videos watched on YouTube. YouTube updated its recommendation algorithm in July 2019, generating significative alterations for the channels of independent content creators. YouTube does not specify how they have modified the recommendation system or what will be the consequences, but many of the independent content creators themselves quickly noticed how these changes affected the traffic on their channels. However, there is still no evidence verifying if independent content creators in YouTube have modified their usual behavior on the platform. This paper analyses if there are changes in upload frequency and video duration of the top three YouTubers in Spain and three independent content creators, relying on web scraping techniques and obtaining data from the YouTube API. Preliminary findings show that independent youtubers have progressively reduced the number of videos they upload per week, while the changes do not seem to affect large YouTube channels.
<div>

**Research Question and Hypothesis**

**Main RQ:**
    
Have independent content creators changed their behaviour in their channels after YouTube changed the algorithm?

**Hypotheses**

H1: Independent content creators have decreased their upload frequency after the algorithm changed

H2: Large channels have not modified their upload frequency after the algorithm changed

**Data and Methods**

<div class=text-justify>
For this study we first created an original dataset relying on web scraping techniques and collecting data from the YouTube API. We found a ranking of the top youtubers in Spain and scraped their channel names into a list that we later on used to query the API. We also collected data for the three manually selected medium-range youtubers in order to be able to compare them. After collecting the data and cleaning the dataset, we calculated the upload frequency and average duration of videos for both groups and plotted the results.
<div>

**Results**

Figure 1 shows there is a downward tendency in terms of upload frequency for independent content creators after the algorithm change in July 2019. On the contrary, Figure 2 shows that large channels seem to either keep in a similar trend as before the changes or even upload more frequently than before.

**Limitations**

<div class=text-justify>
The results of this analysis are limited to the data we were able to extract given the quota limitations on the API and the time restrictions that we faced. Future research should take this preliminary diagnosis and replicate our analyses for a larger dataset. Additionally, statistical techniques would be necessary to determine if the modifications we have detected are significant.
<div>

**Conclusion**

These preliminary findings suggest that independent content creators may be considering alternative platforms or sources of income given that YouTube has modified the terms of their work. However, large channels might grow on the platform since they do not seem to be disfavored by the changes. Further research is needed to confirm this trends.
