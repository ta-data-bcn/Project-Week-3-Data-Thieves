Import all relevant modules

```python
import datetime
import pandas as pd
```

Import the database

```python
top_3=pd.read_csv(r'../Data/Top3_cleaned.csv', sep="\t")
```

```python
top_3.head(10)
```

***Data Cleaning***

The csv generates string objects

```python
top_3.dtypes
```

Transform the dates from strings to datetime objects

```python
top_3.Published = pd.to_datetime(top_3.Published)
```

```python
top_3.dtypes
```

Get the week number for each date and append a column with the week number

```python
top_3["Week_N"]=top_3.Published.apply(lambda el: el.isocalendar()[1])
top_3.head(10)
```

Get the count of videos per week

```python
perweek=top_3.groupby("Week_N").count()
```
