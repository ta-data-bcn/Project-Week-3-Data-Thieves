```python
f = pd.read_csv('animals_population_temperature.csv')

df = f[['Year','Subtotal\n(Vertebrates)','Subtotal\n(Invertebrates)', 'Subtotal\n(Plants)','TOTAL']]


l_total = []
l_vert = []
l_inv = []
l_plants = []
for i in df['Year']:
    a,t = df[df['Year']==i]['TOTAL']
    l_total.append(t/a)
    
    a,t = df[df['Year']==i]['Subtotal\n(Invertebrates)']
    l_inv.append(t/a)

    a,t = df[df['Year']==i]['Subtotal\n(Plants)']
    l_plants.append(t/a)
    
    
    a,t = df[df['Year']==i]['Subtotal\n(Vertebrates)']
    l_vert.append(t/a)
    

f['Index_total'] = l_total
f['Index_vert'] = l_vert
f['Index_inv'] = l_inv
f['Index_plants'] = l_plants


f.to_csv('apt_full_indexed.csv', index=False)
```
