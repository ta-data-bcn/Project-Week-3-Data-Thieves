```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('animals_population_temperature.csv')

data
```

```python
data.columns
```

```python
cols = ['Year', 'Assessment', 'Subtotal\n(Vertebrates)','Subtotal\n(Invertebrates)','Subtotal\n(Plants)', 'TOTAL',
       'Population, total', 'Lowess(5)']

data_reduced = data[cols]

data_reduced.dropna()
```

```python
data_assessed = data_reduced[data_reduced['Assessment'] == 'Total assessed'].set_index('Year')
data_threatened = data_reduced[data_reduced['Assessment'] == 'Total threatened'].set_index('Year')
```

```python
data_assessed.drop(columns=['Assessment'])
data_threatened.drop(columns=['Assessment'])

data_assessed.to_csv('assessed_data.csv')
data_threatened.to_csv('threatened_data.csv')
```
