# Data cleaning

```python
#clean unnecessary columns
br_ppl_pd = br_ppl_pd.drop(columns=['countryiso3code', 'unit', 'obs_status', 'decimal', 'indicator.id', 'country.id'])
br_ppl_pd
```

```python
#clean unnecessary columns
br_migrants_pd = br_migrants_pd.drop(columns=['countryiso3code', 'unit', 'obs_status', 'decimal', 'indicator.id', 'country.id'])
br_migrants_pd
```

```python
#verifying data types
br_migrants_pd.info()
```

```python
#changing data types
br_migrants_pd['date'] = br_migrants_pd['date'].astype('int')
br_ppl_pd['date'] = br_migrants_pd['date'].astype('int')
```

```python
#verifying changes
br_migrants_pd.info()
```

# Filtering data

```python
#obtaining values by Year range
bins = pd.cut(br_migrants_pd['date'], [1960, 1970, 1980, 1990, 2000, 2010, 2019])
migrants_year_range = br_migrants_pd.groupby(bins)['value'].sum()
print(migrants_year_range)
```

```python
#obtaining values by Year range
ppl_year_range = br_ppl_pd.groupby(bins)['value'].sum()
print(ppl_year_range)
```

```python
#create a DataFrame for the new filtered data
migrants_df = pd.DataFrame(migrants_year_range)
ppl_df = pd.DataFrame(ppl_year_range)
demogr_df = pd.concat([migrants_df, ppl_df], axis=1)
demogr_df.columns = ['migrants', 'population']
demogr_df.info()
```

```python
demogr_df['migrants'] = demogr_df['migrants'].astype('int')
```
